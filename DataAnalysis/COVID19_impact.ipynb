## 데이터 분석 연습을 위한 개인 학습 과정

```python
import pandas as pd 
data_frame = pd.read_csv('Data1.txt',encoding='utf-8-sig',delimiter='\t')
print(data_frame.shape) # 행과 열 갯수 확인 
```

```python
data_frame
```

먼저, 데이터들을 분석하기에 앞서 9021개의 데이터들 중 코로나의 심각도 단계에 맞게 그룹별로 데이터를 분류 해준다. 

```python
covid19_phase1 = data_frame[data_frame['Severity_Group']=='COVID19_Phase1']
covid19_phase2 = data_frame[data_frame['Severity_Group']=='COVID19_Phase2']
covid19_phase3 = data_frame[data_frame['Severity_Group']=='COVID19_Phase3']
covid19_phase4 = data_frame[data_frame['Severity_Group']=='COVID19_Phase4']
healthy = data_frame[data_frame['Severity_Group']=='Healthy']
```

분류한 데이터들 즉, 코로나 1 2 3 4단계와 건강한 사람이 9012명 중 몇 명씩 있는지 체크 해본다  
이 데이터 상 감염율이 약 49.939% 인 것을 알 수 있다

```python
print(covid19_phase1.shape)
print(covid19_phase2.shape)
print(covid19_phase3.shape)
print(covid19_phase4.shape)
print(healthy.shape)
H = round(len(healthy)/len(data_frame)*100,3)
print('정상인 비율 = ',H,'%')
print('코로나 확진자 비율 =',100-H,'%')
```

앞서 추출한 정상인 , 1단계 , 4단계 세 그룹의 데이터들을 비교해서 분석해보자 


리스트를 이용해서 비교할 수치들의 이름들을 다 넣어두고 각 그룹의 describe 결과의 1번 인덱스인 mean 에 접근해서  
비교할 모든 수치에 대해서 비교를 진행해 본다

```python
norm = ['Age','Height_CM','Weight_KG','sysBP','HR','Resting_SaO2','BMI','FEV1pp_utah','FVCpp_utah','FEV1_FVC_utah']
```

```python
# Healthy , 1단계 , 4단계 환자의 수치 비교 

for value in norm:
    print(value,'\n',pd.Series([healthy.describe()[value][1],covid19_phase1.describe()[value][1],covid19_phase4.describe()[value][1]]
                          ,index=['healthy','phase1','phase4']),'\n')
```

해당 데이터 수치를 확인해보면 FEV1pp_utah, FVCpp_utah ,FEV1_FVC_utah  수치에서  
코로나 4단계와 코로나 1단계,정상인 두 그룹 사이에서 큰 차이를 보이고 있다 

이번에는 세 그룹을 data frame 으로 합쳐버려서 한 눈에 비교해보자

```python
data_healthy_phase1_phase4 = pd.concat([healthy,covid19_phase1,covid19_phase4])
data_healthy_phase1_phase4
```

위에서 구했던 데이터를 Dataframe 으로 만들어서 더 보기 쉽게 만들었다

```python
data_group = data_healthy_phase1_phase4.groupby('Severity_Group').mean() # 여기서 sum() 도 가능 ( 보통 확진자의 수였다면 사용했을 듯)
data_group
```

즉, 코로나의 심각도를 알기 위해서는 FEV1pp_utah, FVCpp_utah ,FEV1_FVC_utah 세 수치가 중요하다는 것을 알 수 있다  
이번에는 수치 간의 상관 관계를 가볍게 훑어보자

```python
covid19_phase1.corr()
```

코로나 1단계의 경우에는 FEV1pp_utah 과 FVCpp_utah 이 양의 선형 상관관계(0.75)에 치우쳐서 있다  
그리고 FEV1_FVC_utah 와 FVCpp_utah 사이에는 음의 선형 상관관계(-0.47) 성향이 드러나는 것을 확인 할 수 있다  
그 외 수치 사이에는 상관관계가 거의 없다

```python
covid19_phase4.corr()
```

코로나 4단계의 경우에는 FEV1pp_utah 과 FVCpp_utah 이 양의 선형 상관관계(0.53)에 치우쳐서 있다  
그리고 FEV1_FVC_utah 와 FVCpp_utah 사이에는 음의 선형 상관관계(-0.53) 성향이 드러나는 것을 확인 할 수 있다  
그 외 수치 사이에는 상관관계가 거의 없다

위 과정을 정리해보자
1. 코로나 심각도를 구분해주는 수치는 FVCpp_utah , FEV1pp_utah , FEV1_FVC_utah 세 개가 있다   
2. 해당 수치들은 FVCpp_utah 를 기준으로 FEV1pp_utah 과는 양의 상관관계를 , FEV1_FVC_utah 과는 음의 상관 관계를 보인다  

단어 의미
* FVC : 최대로 숨을 들이쉰 다음, 최대로 내쉬었을 때의 공기량
* FEV1 : 1초 간 숨을 얼마나 빨리 내쉴 수 있는지
* FEV1_FVC : 기관지폐쇄 유무로 0.7 이하인 경우 기도폐쇄의 정도를 나타낸다 

즉, 코로나의 단계가 심각해 질수록 FVC와 FEV1 수치는 점 점 떨어지고 폐 기능에 주를 둔 질병임을 확인 할 수 있다
